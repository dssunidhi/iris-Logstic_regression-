```python
#import necessary libraray
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#load dataset
df=pd.read_csv("Iris.csv")
df.head()
```

```python
#species values in object type want to transform so drop one value and convert rest in 0 and 1 format
df=df[df['Species']!='Iris-setosa']
df.head()
```

```python
#check unique values
df['Species'].unique()
```

```python
#value of species replace with 0 and 1
df['Species'] = df['Species'].map({'Iris-versicolor': 0, 'Iris-virginica': 1}).fillna(0).astype(int)
```

```python
df.head()
```

```python
#split dataset into two parts
x=df.iloc[::,0:5]
x
y=df.iloc[::,5]
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
regression=LogisticRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```
